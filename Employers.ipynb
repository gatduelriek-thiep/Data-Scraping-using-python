```python
# Import BeautifulSoup library for HTML parsing
from bs4 import BeautifulSoup
# Import requests library for making HTTP requests
import requests

# Define the URL of the Wikipedia page we want to scrape
url = "https://en.wikipedia.org/wiki/List_of_largest_employers"
# Define headers with a user-agent to make the request more browser-like
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
# Send an HTTP GET request to the URL with headers and store the response
page = requests.get(url, headers=headers)
# Parse the HTML content of the page using BeautifulSoup with the HTML parser
soup = BeautifulSoup(page.text, "html")
```

```python
# print(soup.prettify())
```

```python
#soup.find_all("table")[0]
```

```python
table = soup.find_all("table")[0]
```

```python
world_titles = table.find_all("th")
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_Data = table.find_all("tr")
for row in column_Data[1:]:
    row_data = row.find_all("td")
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv(r"Desktop\Gatduel.csv", index = False)
```
