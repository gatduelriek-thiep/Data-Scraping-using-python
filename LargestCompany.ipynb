## Gatduel is scraping the list of largest Company in USA by Revenue

```python
# Import the modules and Libraries here. I am going to use BeautifulSoup, Requests, and Pandas

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"}
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
page = requests.get(url, headers = headers)

soup = BeautifulSoup(page.text, "html.parser")
table = soup.find_all("table")[0]
```

```python
world_titles = table.find_all("th")
```

```python
world_titles_table = [title.text.strip() for title in world_titles]
```

```python
world_titles_table
```

```python
df = pd.DataFrame(columns = world_titles_table)
```

```python
df
```

```python
column_data = table.find_all("tr")
```

```python
for row in column_data[1:]:
    row_data = row.find_all("td")
    individual_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_data
```
